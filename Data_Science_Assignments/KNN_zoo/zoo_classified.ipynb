```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
cv_scores = []

for i in range(1,30,3):
    knn_model = KNeighborsClassifier(n_neighbors = i)
    cross_val_scores = cross_val_score(estimator=knn_model,X=X_scaled,y=y,cv=5)
    cv_scores.append(cross_val_scores.mean())
```

```python
plt.plot(range(1,31,3),cv_scores)
plt.title('neighbors vs cvscore')
plt.xlabel('neighbors')
plt.ylabel('mean accuracy')
plt.show()
```
