```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
 bike_details = pd.read_csv('mtcars.csv')
```

```python
bike_details
```

```python
bike_details.groupby(by='am')['mpg'].mean()
```

```python
bike_details.groupby(by='cyl')['mpg'].mean()
```

```python
bike_details.groupby(by=['am','cyl'])['mpg'].mean()
```

```python
pd.crosstab(bike_details['cyl'],bike_details['gear'],margins=True)
```

```python
plt.hist(x = 'mpg', data=bike_details)
plt.xlabel('MPG')
plt.title('MPG Distribution')
plt.show()
```

```python
plt.hist(x = 'am', data=bike_details)
plt.xlabel('AM')
plt.title('AM Distribution')
plt.show()
```

```python
bike_details['gear'].value_counts()
```

```python
import seaborn as sns
sns.countplot(x='gear',y=None,hue=None,data=bike_details)
plt.title('Gear distribution')
plt.show
```

```python
sns.histplot(data=bike_details,x='mpg')
```

```python
bike_details['cyl'].value_counts()
```

```python
plt.figure(figsize=(10,10))
plt.pie(x=bike_details['cyl'].value_counts(),labels=[8,4,6],explode=[0.02,0.02,0],autopct='%.2f',)
plt.title('percentage of cylinders')
plt.show()
```

```python
sns.scatterplot(x='hp',y='mpg',data=bike_details)
plt.title('HP vs MPG')
plt.show()
```

```python
sns.barplot(x='gear',y='mpg',data=bike_details)
plt.show
```

```python
sns.pairplot(data=bike_details)
plt.show
```

```python
corr_matrix = bike_details.corr().round(2)
corr_matrix
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(data = corr_matrix,annot=True)
plt.show()
```

```python
plt.boxplot(x='hp',data = bike_details)
plt.show()
```

```python
sns.boxplot(x='gear',y='mpg',data=bike_details)
plt.show
```
