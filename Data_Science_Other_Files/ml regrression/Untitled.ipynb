## use statsmodels for lr

```python
x = cars_data.drop(labels='MPG',axis=1)
```

```python
import statsmodels.formula.api as smf
lin_mod_1 = smf.ols(formula = 'MPG ~ HP + VOL + SP + WT', data = cars_data ).fit()
y_pred = lin_mod_1.predict(x)
y_pred
```

```python
lin_mod_1.params
```

```python
lin_mod_1.pvalues
```

```python
print('r2 score            :' , round(lin_mod_1.rsquared,5))
print('adjustred r2 score  :' , round(lin_mod_1.rsquared_adj,5))
print('AIC                 :' , round (lin_mod_1.aic,5))
print('BIC                 :' , round(lin_mod_1.bic,4))
```
