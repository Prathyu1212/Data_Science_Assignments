```python
import pandas as pd
```

```python
from sklearn import datasets
```

```python
 breast_cancer_data = datasets.load_breast_cancer()
```

```python
print(breast_cancer_data.DESCR)
```

```python
cancer_data = pd.DataFrame(data=breast_cancer_data.data,columns=breast_cancer_data.feature_names)
cancer_data['target'] = breast_cancer_data.target
cancer_data
```

```python
cancer_data.columns
```

```python
cancer_data.shape
```

```python
cancer_data.isna().sum()
```

```python
cancer_data.dtypes
```

```python
x = cancer_data.drop(labels=['target'],axis=1)
y= cancer_data[['target']]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
train_test_split(x,y,test_size=0.20,random_state=12,shuffle=True,stratify=y)
```
