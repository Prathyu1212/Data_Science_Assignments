```python
user_details.groupby(by='occupation')['user_id'].count().sort_values(ascending = False) #aggregation
```

```python
user_details.groupby(by=['gender','occupation'])['user_id'].count().sort_values(ascending = False) #aggregation
```

```python
pd.pivot_table(data = user_details,values='user_id',index='occupation',columns='gender',aggfunc='count')
```

```python
pd.set_option('max_columns',None) #no restrictions for your columns
```

```python
user_details.groupby(by = ['gender','occupation'])['age'].mean().round()
```

```python
pd.pivot_table(data = user_details,values='age',index='occupation',columns='gender',aggfunc='mean').round()
```
